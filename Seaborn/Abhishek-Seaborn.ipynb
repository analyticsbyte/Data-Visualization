```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
dataset = pd.read_csv('Real_Estate.csv')
```

```python
dataset.head()
```

# **** Distribution Plot - Neumerical Data ****

#### Total distribution of one variable to other 

```python
sns.distplot(dataset['price'])
```

```python
sns.distplot(dataset['price'], kde = False, bins  =30) 

#its a Histogram
```

### Joint Plot

#### This plot helps to merge two plot together and check its distribution

```python
sns.jointplot(x='price',y='size',data = dataset)
```

```python
sns.jointplot(x='price',y='size',data = dataset, kind = 'hex') # hexagon distribution
                                                              # more points gets darker and less point goes lighter
```

```python
sns.jointplot(x='price',y='size',data = dataset, kind = 'reg') # linear fit , regression line
```

## Pairplot

Joint plot of every single numerical coloumns in this data frames

```python
sns.pairplot(dataset) # scatterplot for other varable and histogram for same
```

# Catagorical Plots

```python
#dataset_pur = pd.read_csv('Purchased_Data.csv')
data_tips = sns.load_dataset('tips')
```

```python
data_tips
```

```python
data_tips.info()
```

### Bar Plot

```python
plt.figure(figsize=(10, 6)) # increase the size of the figure
sns.barplot(x='sex',y='total_bill', data = data_tips)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='sex',y='total_bill', data = data_tips, estimator= np.std)
```

### Count Plot

Similar as barplot but only count number of occurences in the dataset

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', data = data_tips) # number of occurences of  male and female in the dataset
```

# **** Distribution Plot - catagorical data ****

### Box Plot - one catagorical & another neumerical

```python
plt.figure(figsize=(10,6))

sns.boxplot(x='day', y= 'total_bill', data = data_tips)
```

```python
plt.figure(figsize=(16,6))

sns.boxplot(x='day', y= 'total_bill', data = data_tips, hue = 'smoker')
```

# **** Matrix Plot - Heat Map ****

```python
dataset
```

Correlation is a statistical term which in common usage refers to how close two variables are to having a linear relationship with each other.

Features with high correlation are more linearly dependent and hence have almost the same effect on the dependent variable. So, when two features have high correlation, we can drop one of the two features.

```python
dataset.corr()
```

```python
sns.heatmap(dataset.corr())
```

```python
sns.heatmap(dataset.corr(), annot = True) #annotations
```

```python
sns.heatmap(dataset.corr(), cmap = 'magma',linecolor = 'red', linewidths =3) 
```
